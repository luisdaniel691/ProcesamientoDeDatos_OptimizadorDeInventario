**Datos de entrada**

```python
# Formato: (ID_Producto, Categoría, Precio, Stock_Actual)
entradas_almacen = [
    ("PROD-001", "Electrónica", 150.0, 5),
    ("PROD-002", "Hogar", 25.0, 10),
    ("PROD-001", "Electrónica", 150.0, 5),  # Duplicado
    ("PROD-003", "Ropa", 12.0, 0),
    ("PROD-004", "Electrónica", 300.0, 2),
    ("PROD-002", "Hogar", 25.0, 10),        # Duplicado
    ("PROD-005", "Ropa", 45.0, 15)
]
```

**1. Depuración de Inventario (Sets)**

```python
productos_unicos = set(entradas_almacen)
```

**Pregunta técnica: ¿Por qué podemos convertir una lista de tuplas a un set sin errores, pero no podríamos hacerlo si fuera una lista de listas? (Pista: Inmutabilidad).**

Porque los sets solo pueden tener elementos inmutables, o tambien conocidos como hashables. Las tuplas una vez creadas se vuelven inmutables, pero las listas son mutables porque se les pueden quitar o agregar elementos en cualquier momento.
Por lo tanto, como los sets solo aceptan elementos inmutables, si podria convertir una lista de tuplas a un set, pero no lo podria hacer con una lista de listas (porque estas son mutables).

**Intenta cambiar la latitud a 20.0 y captura el error (o coméntalo) explicando por qué esto protege la integridad de los datos de geolocalización.**

En la siguiente linea intentamos cambiar su valor de latitud a 20, pero nos da un error, donde dice: TypeError: 'tuple' object does not support item assignment.

El error es muy claro, dice que no permite asignacion de valor por que es un objeto de tipo tupla, y la naturaleza de las tuplas es ser inmutable, es decir, que una vez creada una tupla y asignandole valores, esos valores ya no pueden ser cambiados, por eso es que nos sale ese error, porque estamos intentando cambiarle uno de sus valores a una tupla.

**3. Filtro de Reposición (Listas)**

```python
pedidos_urgentes = []
for i in productos_unicos:
  if i[3] == 0:
    pedidos_urgentes.append(i[0])
print(pedidos_urgentes)
```

**4. Dashboard de Categorías (Diccionarios)**

```python
categorias_unicas = set()
for i in productos_unicos:
  categorias_unicas.add(i[1])

resumen_almacen = {
    "total_items": len(productos_unicos),
    "categorias": categorias_unicas,
    "alerta_stock": len(pedidos_urgentes) > 0
}
print(resumen_almacen)
```
